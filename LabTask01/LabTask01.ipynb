### Part 3: Filtering ###

1. Show only the rows where the age is above 60.

2. Show rows where the fare is above 100 and the passenger was in 1st class.

3. Identify and isolate rows with missing values in the age column.

I filtered the data to show only passengers over 60 and first-class passengers who paid a fare over 100. I also used .isna() to find the rows where the age is missing.

```python
# 1. Passengers older than 60
df[df['Age'] > 60]

# 2. Fare > 100 AND 1st class — each condition in parentheses, use & not "and"
df[(df['Fare'] > 100) & (df['Pclass'] == 1)]

# 3. Rows with missing Age
df[df['Age'].isna()]
```

### Part 4: Reindexing and Dropping ###

1. Reverse the row order of the dataset.

2. Remove one column from the dataset.

3. Remove all rows with missing values in the age column.

I reversed the order of the rows and removed the Cabin column entirely. I also used .dropna() to create a version of the data where no ages are missing.

```python
# 1. Reverse row order
df_reversed = df.iloc[::-1]
df_reversed.head()

# 2. Drop a column (e.g., Cabin — it has lots of missing values anyway)
df_dropped = df.drop('Cabin', axis=1)
df_dropped.head()

# 3. Drop all rows where Age is missing
df_no_missing_age = df.dropna(subset=['Age'])
print("Before:", df.shape, "After:", df_no_missing_age.shape)
```

### Part 5: Arithmetic and Data Alignment ###

1. Create a new column that converts fares from USD to Euros.

2. Subtract the average age from every age value.

3. Subtract a custom array from two selected columns in the dataset.

I created a new column to convert the fare to Euros. I also subtracted the average age from all age values and used a custom array to subtract values from the age and fare columns.

```python
# 1. Fare USD → Euros (assume 1 USD ≈ 0.92 EUR)
df['Fare_EUR'] = df['Fare'] * 0.92

# 2. Subtract the average age from every age value (centering)
df['Age_centered'] = df['Age'] - df['Age'].mean()

# 3. Subtract a custom NumPy array from two columns
custom = np.array([1, 2])  # subtract 1 from Age, 2 from Fare for each row
two_cols = df[['Age', 'Fare']]
result = two_cols.sub(custom, axis=1)   # axis=1 aligns array along columns
result.head()
```

### Part 6: Applying Functions ###

1. Create a new column with the number of characters in each name.

2. Calculate the difference between fare and age for each row.

3. Round all numerical values in the dataset to one decimal place.

I used the apply method to count the characters in each name and to find the difference between the fare and age. I also rounded all numbers in the dataset to one decimal place.



```python
# 1. Character count of each name
df['Name_length'] = df['Name'].apply(len)

# 2. Fare − Age for each row (row-wise apply)
df['Fare_minus_Age'] = df.apply(lambda row: row['Fare'] - row['Age'], axis=1)

# 3. Round all numeric values to 1 decimal place
df_rounded = df.round(1)
df_rounded.head()
```

### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).

I mapped the gender strings to numbers and created a function to group passengers into age categories. I made the function identify missing ages as Unknown so that the data stays accurate.

```python
# 1. Map gender to numeric codes
df['Sex_code'] = df['Sex'].map({'male': 0, 'female': 1})

# 2. Categorize age into groups
def age_group(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['Age_group'] = df['Age'].map(age_group)
df[['Age', 'Age_group', 'Sex', 'Sex_code']].head(10)
```

# Part 8: Performance Comparison — Fare Conversion #

In this section, I compare the execution time of **Part 5 Task 1** (converting Fares from USD to EUR). 

I compared the time it takes to convert the Fare column from USD to EUR using a pure Python loop versus the Pandas method. This shows the difference in execution speed between the two

```python
import csv
import urllib.request
import time

# 1. Prepare Pure Python Data (List of Dictionaries)
url = "https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv"
with urllib.request.urlopen(url) as response:
    raw_data = response.read().decode('utf-8').splitlines()

reader = csv.DictReader(raw_data)
python_data = list(reader)

# 2. Benchmark Pure Python
start_py = time.perf_counter()
fares_eur_py = [float(row['Fare']) * 0.92 if row['Fare'] != '' else 0.0 for row in python_data]
end_py = time.perf_counter()

# 3. Benchmark Pandas
start_pd = time.perf_counter()
fares_eur_pd = df['Fare'] * 0.92
end_pd = time.perf_counter()

# 4. Results Table
t_py = end_py - start_py
t_pd = end_pd - start_pd

print(f"{'Method':<20} | {'Time (seconds)':<15}")
print("-" * 38)
print(f"{'Pure Python':<20} | {t_py:<15.6f}")
print(f"{'Pandas':<20} | {t_pd:<15.6f}")

if t_pd < t_py:
    print(f"\nResult: Pandas was {t_py/t_pd:.2f}x faster.")
else:
    print(f"\nResult: Pure Python was {t_pd/t_py:.2f}x faster")
```
